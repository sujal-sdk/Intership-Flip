# Question 2 Naukri.com

### Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name. You have to scrape first 10 jobs data.

```python
#program to open website
url='https://www.naukri.com/'
driver.get(url)

#program to find searchbar
search_jobs=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[1]/div/div/div/input')
search_jobs

#program to enter words in searchbar
search_jobs.send_keys('Data Scientist')

#program to find searchbar
search_loc=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
search_loc

#program to enter words in searchbar
search_loc.send_keys('Bangalore')

#program to find search button
search_btn=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_btn

#program to click on search button
search_btn.click()
```

```python
#program to extract all xpaths
title_tags=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
len(title_tags)
title_tags=title_tags[0:11]

#program to find job titles
job_titles=[]
for i in title_tags:
    job_titles.append(i.text)
job_titles
#program to extract all xpaths
company_tags=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
len(company_tags)

company_tags=company_tags[0:11]

#program to find company names
comp_name=[]
for i in company_tags:
    comp_name.append(i.text)
comp_name

#program to extract all xpaths
exp_tags=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
len(exp_tags)

exp_tags=exp_tags[0:11]

#program to find experience
exp=[]
for i in exp_tags:
    exp.append(i.text)
exp

#program to extract all xpaths
loc_tags=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
len(loc_tags)

loc_tags=loc_tags[0:11]

#program to find location
loc=[]
for i in loc_tags:
    loc.append(i.text)
loc

#program to create a DataFrame
df=pd.DataFrame({'Job Title':job_titles,'Company':comp_name,'Experience':exp,'Location':loc})
df
```

```python
#program to find 100 brand description
desc=[]
for page in range(0,2,1):
    desc_tag=driver.find_elements_by_xpath('//h4[@class="product-product"]')
    for i in desc_tag:
        desc.append(i.text)
print(len(desc))
print(desc)
```

```python
#program to create a database
df=pd.DataFrame({'Brand':brand, 'Shoe Description':desc, 'Price':price})
df
```
